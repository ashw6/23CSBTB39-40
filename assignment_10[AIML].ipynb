```python
import sklearn
from sklearn.model_selection import train_test_split

X = power_com_df.drop(['Date','Time'], axis=1)
y = power_com_df.Voltage

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
X_train.head()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np

kmeans = KMeans(n_clusters=3)
for col in ['Global_active_power', 'Global_reactive_power']:
    X_train[col] = pd.to_numeric(X_train[col], errors='coerce')
X_train = X_train.dropna(subset=['Global_active_power', 'Global_reactive_power'])
kmeans.fit(X_train[['Global_active_power', 'Global_reactive_power']])
```
